```python
# Filter data for female population
female_population=population[population['Indicator Code'] == 'SP.POP.TOTL.FE.IN']

# Sort data based on the female population for 2022
female_population_sort=female_population.sort_values(by=2022 , ascending=False)

# Get the top fifteen countries with the highest female population for 2022
female_population_top=female_population_sort.head(15)
female_population_top
```

```python
# Filter data for male population
male_population=population[population['Indicator Code']=='SP.POP.TOTL.MA.IN']

# Sort data based on the male population for 2022
male_population_sort=male_population.sort_values(by=2022 , ascending=False)

# Get the top fifteen countries with the highest male population for 2022
male_population_top=male_population_sort.head(15)
male_population_top
```

```python
plt.figure(figsize=(30,30))
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
#female
plt.subplot(2,1,1)
plt.title('Top fifteen Countries with Highest Female Population (2022)',size=30)
sns.barplot(x=2022,y='Country Name' , data=female_population_top,palette='bright')
plt.xlabel('Female Population',size=25)
plt.ylabel('Countries', size=25)
plt.xticks(size=30)
plt.yticks(size=30)

#male
plt.subplot(2,1,2)
plt.title('Top fifteen Countries with Highest Male Population (2022)',size=30)
sns.barplot(x=2022,y='Country Name' , data=male_population_top,palette='bright')
plt.xlabel('Male Population',size=25)
plt.ylabel('Countries',size=25)
plt.xticks(size=30)
plt.yticks(size=30)

plt.show()
```

```python
# Merge male and female population data on 'Country Name'
merged_population=pd.merge(male_population,female_population,on='Country Name' , suffixes=('_male','_female'))

# Calculate the total population for each country (male + female)
merged_population['Total Population']=merged_population['2022_male'] + merged_population['2022_female']

# Sort data based on total population in ascending order
total_merged_population=merged_population.sort_values(by='Total Population',ascending=False)

# Select the top 15 countries with the highest total population
merged_population_top=total_merged_population.head(15)
merged_population_top
```

# Visualizing The Data

```python
#stacked bar plot
plt.figure(figsize=(13,5))
sns.set(style='whitegrid')

plt.title('Top 15 Countries with Male and Female Populations (2022)')
sns.barplot(x='Country Name' , y='2022_male' , data=merged_population_top , color='#1C1C1BFF' , label='Male population' )
sns.barplot(x='Country Name' , y='2022_female' , data=merged_population_top , bottom=merged_population_top['2022_male'] ,label='Female population' , color='#CE4A7EFF')
plt.legend()
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45,ha='right')
plt.show()
```
